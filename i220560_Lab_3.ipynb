## Email Classification
Develope a tool which will classify an email.

### Tasks:
* Create a function classify_emails that takes a list of emails as an argument.
* Inside the function, categorize emails into "Important," "Promotions," and "Spam" based on keywords and content analysis.
* Loop through the emails and identify keywords to determine their category.
* If an email contains keywords indicating importance, assign it as "Important."
* If an email contains keywords related to promotions or deals, assign it as "Promotions."
* If an email is suspected to be spam based on certain keywords or patterns, assign it as "Spam."
* Return three lists of emails for each category.
* Print the lists.

```python
from nltk.corpus import wordnet

def get_syns(words):
    result = []
    for word in words:
        synonyms = []
        for syn in wordnet.synsets(word):
            for lemma in syn.lemmas():
                synonyms.append(lemma.name())
        result.extend(synonyms)
    return set(result)
```

```python
def classify_mails(emails):

  prm = ["sale", "deal", "discount", "promotion", "special offer", "exclusive", "clearance", "new arrival", "limited time", "bonus", 'claim']
  spm = ["free", "prize", "discount", "win", "limited time", "money", "click here", "act now", "offer", "guaranteed"]
  ip = ["urgent", "important", "action required", "deadline", "critical", "response needed", "attention", "priority", "confidential", "emergency"]

  promo = get_syns(prm)
  spam = get_syns(spm)
  imp = get_syns(ip)
  result = []
  for email in emails:
    pword_count = 0
    iword_count = 0
    sword_count = 0

    for word in word_tokenize(email):
      if word in promo:
        pword_count += 1
      elif word in spam:
        sword_count += 1
      elif word in imp:
        iword_count += 1

    max_dict = {'Promotion': pword_count, 'Spam': sword_count, 'Important': iword_count}
    result.append(max(max_dict, key=max_dict.get))
    # print('mmmm',max_dict)
  return result



```

```python

i = """
We hope this email finds you well. We wanted to remind you about the upcoming conference scheduled for next week.
Your presence is crucial, and we need your confirmation as soon as possible.
Please take a moment to confirm your attendance by clicking on the following link: [Confirmation Link].
If you have any questions or concerns, feel free to reach out to us.
Thank you for your prompt attention to this matter.
"""
p = """
Exciting news! We're thrilled to announce a special promotion exclusively for our loyal customers.
For a limited time only, enjoy a 50% discount on all our new arrivals.
Hurry, this offer won't last long! Visit our website or stop by our store to explore our latest collection.
Don't miss out on this fantastic opportunity to refresh your wardrobe for less.
"""
s = """
We are thrilled to inform you that you've been selected as the grand prize winner of our exclusive sweepstakes. You have won a whopping $1,000,000!
To claim your prize, simply click on the link below and provide your personal information. Act now to secure your winnings!
Claim Your Prize Now: [Claim Link]
Congratulations again on your incredible win!
"""
rs = classify_mails([i,s,p])
rs
```

## English-Spanish Dictionary
You are creating a basic English-Spanish dictionary that allows users to look up English words and find their Spanish translations.


### Tasks
* Create a dictionary where English words are keys and their Spanish translations are values.
* Ask the user to input an English word.
* Check if the word exists in the dictionary.
* If the word exists, print its Spanish translation.
* If the word does not exist, inform the user that the translation is not available.
* Implement a loop that allows the user to keep looking up words until they choose to exit.


```python
#code here
dct = {'hello': 'hola', 'goodbye': 'adiós', 'thank you': 'gracias', 'yes': 'sí', 'no': 'no', 'cat': 'gato', 'dog': 'perro', 'house': 'casa', 'friend': 'amigo', 'love': 'amor'}
exit = 'q'
while True:
  inp = input('Please enter an English word for Spanish translation (press q to quit): ')
  inp = inp.lower()
  if inp == exit:
    break
  eng = dct.keys()
  if inp in eng:
    print(dct[inp])
  else:
    print('Translation is not available')
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/TirendazAcademy/PANDAS-TUTORIAL/main/DataSets/vgsalesGlobale.csv')

global_sum = df['Global_Sales'].sum()
print('global sales sum',global_sum)

global_max = df['Global_Sales'].max()
print('global sales max',global_max)

nu_avg = df['NA_Sales'].mean()
eu_avg = df['EU_Sales'].mean()
jp_avg = df['JP_Sales'].mean()
other_avg = df['Other_Sales'].mean()
print()
print(f"Average NA Sales: {nu_avg}")
print(f"Average EU Sales: {eu_avg}")
print(f"Average JP Sales: {jp_avg}")
print(f"Average Other Sales: {other_avg}")
```

```python

popular_5 = df.nlargest(5, 'Global_Sales')
print('top 5: ' )
popular_5
```

```python
nu_total = df['NA_Sales'].sum()
eu_total = df['EU_Sales'].sum()
jp_total = df['JP_Sales'].sum()
other_total = df['Other_Sales'].sum()
print(f"Total NA Sales: {nu_total}")
print(f"Total EU Sales: {eu_total}")
print(f"Total JP Sales: {jp_total}")
print(f"Total Other Sales: {other_total}")

sales = df.groupby('Genre')['NA_Sales'].sum()
genre_popular = sales.idxmax()
print('popular genre in NA: ', genre_popular)
print()

df.tail(10)
```

```python
global_sales = df.groupby('Platform')['Global_Sales'].sum().nlargest(3)
global_sales
for p, i in global_sales.groupby(level=0):
    print(p)

print()
genre_sales = df.groupby('Genre')['Global_Sales'].sum()
total = df['Global_Sales'].sum()
market_share = (genre_sales / total) * 100
print("Market share\n",market_share)
print()

crltn = df['NA_Sales'].corr(df['EU_Sales'])
print('correlation: ', crltn)
```
